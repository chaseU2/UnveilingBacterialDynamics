```python
import numpy as np
import matplotlib.pyplot as plt

# Time steps
T = 110
time = np.arange(0, T)

# Initialize arrays
y1 = np.zeros(T)
y2 = np.zeros(T)
y3 = np.zeros(T)
y4 = np.zeros(T)
y5 = np.zeros(T)
y6 = np.zeros(T)
y7 = np.zeros(T)
y8 = np.zeros(T)
y9 = np.zeros(T)
y10 = np.zeros(T)
y11 = np.zeros(T)
y12 = np.zeros(T)

# Initial conditions
y1[0] = 0.5
y2[0] = 0.3
y3[0] = 0.6
y4[0] = 0.7
y5[0] = 0.6
y6[0] = 0.4
y7[0] = 0.5
y8[0] = 0.6
y9[0] = 0.7
y10[0] = 0.6
y11[0] = 0.2
y12[0] = 0.4

# Parameters
alpha = 0.02  # damping factor
r = 3.5       # for logistic maps (chaotic interaction)
A = 1.0       # amplitude for nonlinear interaction
B = 0.1       # noise scale for feedback loop
beta = 2.0    # scaling factor for linear relationship in system 6

# Random number generator for noise
rng = np.random.default_rng()

for t in range(T - 1):
    # Small random noise
    noise = rng.normal(0, 0.1, 6)
    
    # System 1: Directed Causation
    y1[t + 1] = np.sin(time[t]) + 0.5 * y1[t] + noise[0]
    if t < 4:
        y2[t + 1] = 0.5 * (y1[t])**2 + 0.2 * y2[t] + 0.4 * noise[1]
    else:
        y2[t + 1] = 0.4 * (y1[t-1])**2 + 0.2 * y2[t] + 0.4 * noise[1]
    
    # System 2: Bidirectional Causation
    y3[t + 1] = 0.9 * y4[t] + 0.1 * np.sin(time[t]) + noise[2]
    y4[t + 1] = 0.9 * y3[t] + 0.1 * np.cos(time[t]) + noise[3]
    
    # System 3: Independent Dynamics
    y5[t + 1] = 0.5 * np.sin(time[t]) + 0.5 * y5[t] + noise[0]
    y6[t + 1] = 0.8 * y6[t] + noise[1]
    
    # System 4: Nonlinear Interaction
    y7[t + 1] = A * np.sin(y8[t]) + noise[1]
    y8[t + 1] = A * np.cos(y7[t]) + noise[2]
    
    # System 5: Feedback Loop
    y9[t + 1] = y9[t] + 0.1 * (y10[t] - y9[t]) + B * noise[3]
    y10[t + 1] = y10[t] - 0.1 * (y10[t] - y9[t]) + B * noise[4]
    
    # System 6: Linear Correlation
    y11[t + 1] = 0.9 * y11[t] + noise[5]  # y11 is a simple autoregressive process
    y12[t + 1] = 0.7 * y11[t] + noise[4]  # y12 is linearly correlated with y11

# Plot the results
plt.figure(figsize=(12, 16))

# Plot each system
plt.subplot(6, 1, 1)
plt.plot(time, y1, label='y1', alpha=0.7)
plt.plot(time, y2, label='y2', alpha=0.7)
plt.title('System 1: Directed Causation')
plt.legend()
plt.grid(True)

plt.subplot(6, 1, 2)
plt.plot(time, y3, label='y3', alpha=0.7)
plt.plot(time, y4, label='y4', alpha=0.7)
plt.title('System 2: Bidirectional Causation')
plt.legend()
plt.grid(True)

plt.subplot(6, 1, 3)
plt.plot(time, y5, label='y5', alpha=0.7)
plt.plot(time, y6, label='y6', alpha=0.7)
plt.title('System 3: Independent Dynamics')
plt.legend()
plt.grid(True)

plt.subplot(6, 1, 4)
plt.plot(time, y7, label='y7', alpha=0.7)
plt.plot(time, y8, label='y8', alpha=0.7)
plt.title('System 4: Nonlinear Interaction')
plt.legend()
plt.grid(True)

plt.subplot(6, 1, 5)
plt.plot(time, y9, label='y9', alpha=0.7)
plt.plot(time, y10, label='y10', alpha=0.7)
plt.title('System 5: Feedback Loop')
plt.legend()
plt.grid(True)

plt.subplot(6, 1, 6)
plt.plot(time, y11, label='y11', alpha=0.7)
plt.plot(time, y12, label='y12', alpha=0.7)
plt.title('System 6: Linear Correlation')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# MI

def calc_MI(bakti_1, bakti_2, amount_bins):

    bakti_1_val = validation_data[bakti_1].values
    bakti_2_val = validation_data[bakti_2].values

    joint_counts = np.histogram2d(bakti_1_val, bakti_2_val, bins = amount_bins)[0]
    joint_probs = joint_counts / np.sum(joint_counts)

    marginal_probs_bakti_1 = np.sum(joint_probs, axis=1)
    marginal_probs_bakti_2 = np.sum(joint_probs, axis=0)

    mutual_info = 0

    for i in range(joint_counts.shape[0]):
        
        for j in range(joint_counts.shape[1]):
            
            if joint_probs[i, j] > 0:

                if marginal_probs_bakti_1[i] > 0 and marginal_probs_bakti_2[j] > 0:
                    
                    mutual_info += joint_probs[i, j] * np.log2(joint_probs[i, j] / (marginal_probs_bakti_1[i] * marginal_probs_bakti_2[j]))

    return(mutual_info)


# create empty matrix
all_baktis = validation_data.columns

num_bacteria = len(all_baktis)

MI_matrix = [[0] * num_bacteria for _ in range(num_bacteria)]


# set amount of bins here
amount_bins = 6

# fill with MI values
for i, name1 in enumerate(all_baktis):

    for j, name2 in enumerate(all_baktis):

       #if i == j: # use this to set selv comparison to zero
            #continue

        MI = calc_MI(name1, name2, amount_bins)
        
        MI_matrix[i][j] = MI


co_occurrence_MI = pd.DataFrame(MI_matrix, columns=all_baktis, index=all_baktis)

print(co_occurrence_MI)
```

```python
# checking convergence

L_range = range(5, 110, 2)
tau = 1
E = 2


Xhat_My, Yhat_Mx = [], [] # correlation list

for L in L_range: 
    
    ccm_XY = ccm(y1, y2, tau, E, L) # define new ccm object # Testing for X -> Y
    ccm_YX = ccm(y2, y1, tau, E, L) # define new ccm object # Testing for Y -> X    
    Xhat_My.append(ccm_XY.causality()[0]) 
    Yhat_Mx.append(ccm_YX.causality()[0]) 
    
print('X->Y r', np.round(Xhat_My[-1], 2), 'p value', np.round(ccm_XY.causality()[1], 4))
print('Y->X r', np.round(Yhat_Mx[-1], 2), 'p value', np.round(ccm_YX.causality()[1], 4))    

print(Yhat_Mx)
# plot convergence. Convergence is necessary to conclude causality
plt.figure(figsize=(5,5))
plt.plot(L_range, Xhat_My, label='$Y$ -> $X$')
plt.plot(L_range, Yhat_Mx, label='$X$ -> $Y$')
plt.xlabel('L', size=12)
plt.ylabel('correl', size=12)
plt.legend(prop={'size': 12}) 
```
